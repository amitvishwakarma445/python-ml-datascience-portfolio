```python
import pandas as pd
```

```python
## find missing value [True= missing value, False=Not missing value]

data.isnull()  ## missing value
```

```python
## missing value of particular column

data['Sales'].isnull()
```

```python
## missing value of particular column

data['Value'].isnull()
```

```python
## find missing value column wise

data.isnull().any(axis=1)  ## gives 1st column in which null value occurs
```

```python
## sum of missing value in every column

data.isnull().sum()
```

```python
display(data)
```

```python
## fill missing value with the '0'

data.fillna(0)
```

```python
data
```

```python
## fill missing value(NAN VALUE) of particular column with the "mean of the its calumn"

## add new column 
data['New_Value'] = data["Value"].fillna(data["Value"].mean())  ## the NAN of "value" column filled with mean()
```

```python
data
```

```python
## rename of particular column (more than one column can be rename also)

data.rename(columns={"Date" : "Sales Date", "Value" : "Values"})  ##  rename(columns={old_name : new_nameame})
```

```python
data
```

```python
## using - apply(funation()) 

## increase the Value of "Values" column by 2 times

data["incre_Value"] = data["Value"].apply(lambda x:x*2)
data.head()
```

```python
## find sum of all values of every Products

every_products_value_sum = data.groupby("Product")["Value"].sum()

display(every_products_value_sum)
```

```python
## find mean of all values of every Products

every_products_value_mean = data.groupby("Product")["Value"].mean()

display(every_products_value_mean)
```

```python
## find mean of all "Sales" of every Products (based on Product and Region Column)

every_products_value_mean = data.groupby(["Product","Region"])["Sales"].mean()

display(every_products_value_mean)
```

```python
## BUT if we wamt to find "sum, mean and count" of Sales on every Products at once

## we will use aggrigate function

grouped_agg = data.groupby("Product")["Sales"].agg(['mean', 'sum', 'count'])

display(grouped_agg)
```

```python
## Merging and joining DataFrames

## create sample of datafarme


df1 = pd.DataFrame({
                    "key":['A',"B","C"],
                    "Value1":[1,2,3]
                    })

df2 = pd.DataFrame({
                    "key":['A',"B","D"],
                    "Value2":[4,5,6]
                    })

display(df1,df2)
```

```python
## inner merging

pd.merge(df1, df2, on="key", how="inner")

### on = common column name
### how = join technique(inner/outer,left,right)
```

```python
## inner merging

pd.merge(df1, df2, on="key", how="outer")

### on = from both table find common column name
### how = join technique(inner/outer,left,right)
```

```python
## inner merging

pd.merge(df1, df2, on="key", how="left")

### on = common column name
### how = join technique(inner/outer,left,right)
```

```python
## inner merging

pd.merge(df1, df2, on="key", how="right")

### on = common column name
### how = join technique(inner/outer,left,right)
```
