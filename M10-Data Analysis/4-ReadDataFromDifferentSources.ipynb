# Read Data From Different Sources

```python
import pandas as pd
from io import StringIO
data = '{"emp_name":"James", "email":"james@gmail.com", "job_profile":[{"title1":"Teaam Lead", "title2":"Sr. Developer"}]}'

df = pd.read_json(StringIO(data))
```

```python
df
```

* JSON (JavaScript Object Notation) files, which are widely used for storing and exchanging data.

```python
## convert to JSON file again

df.to_json(orient='index') ### index based orientation
```

```python
## convert to JSON file again

df.to_json(orient='records') ### index based orientation
```

## pickle file in python

* converting an object in the memory to a byte stream that can be stored as a binary file on disk.
* When we load it back to a Python program, this binary file can be de-serialized back to a Python object.

```python
df_excel
```

```python
## convert df_excel to pickle file

df_excel.to_pickle('df_excel_pickle')
```

```python
## read this pickle file 

pd.read_pickle("df_excel_pickle")
```
