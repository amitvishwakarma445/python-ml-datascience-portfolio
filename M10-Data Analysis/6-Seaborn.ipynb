```python
import seaborn as sns
```

```python
## Basic ploting with seaborn

tips_dataSource = sns.load_dataset('tips')   ## tips- built in record in seaborn to study
display(tips_dataSource)
```

```python
## 1. Create a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips_dataSource, color='red')

plt.title("Scatter plot of Bills vs Tip")
plt.show()
```

```python
## line plot

sns.lineplot(x='size', y='total_bill', data = tips_dataSource)
plt.title("line plot of size VS total_bill")
plt.show()
```

```python
## catagorical plots
## barplot

sns.barplot(x="day", y='total_bill', data=tips_dataSource)
plt.title("bar plot of total bill and day")
plt.show()
```

```python
## box plot

sns.boxplot(x="day", y='total_bill', data=tips_dataSource)
plt.title("box plot of total bill and day")
plt.show()
```

```python
## violin plot

sns.violinplot(x="day", y='total_bill', data=tips_dataSource)
plt.title("violin plot of total bill and day")
plt.show()
```

```python
## Histogram Plot

sns.histplot(tips_dataSource['total_bill'], bins=10)
```

```python
## Histogram Plot

sns.histplot(tips_dataSource['total_bill'], bins=10, kde=True) ### kde = Kernel Density Estimation line/curved_line to the histogram.
```

```python
## KDE plot : Kernel Density Estimation line/curved_line to the histogram.

sns.kdeplot(tips_dataSource["total_bill"], fill=True, color='red', edgecolor='black')
```

## corr() method

* is a table with a lot of numbers that represents how well the relationship is between two columns.
<br>

* The number varies from -1 to 1.
<br>

* 1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.

* 0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

* -0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

* 0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

**What is a good correlation?** It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.

```python
## Heatmap :- like a aggrigation function [agg()]

correlations = tips_dataSource[['total_bill', 'tip', 'size']].corr()
display(correlations)


sns.heatmap(correlations, annot=True, cmap="coolwarm")
# annot = annotaion
# cmap = colormap= ('Blues', 'coolwarm', 'viridis', etc.)
```
