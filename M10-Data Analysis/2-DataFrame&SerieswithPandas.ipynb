## Pandas
* Pandas is a powerful and open-source Python library.
* The Pandas library is used for data manipulation and analysis.
* Pandas consist of data structures and functions to perform efficient operations on data.
* The Pandas library is generally used for data science.
* Pandas is often used as input for plotting functions in Matplotlib, statistical analysis<br> in SciPy, and machine learning algorithms in Scikit-learn.
  
### Data Structures in Pandas Library
Pandas generally provide two data structures for manipulating data. They are:

1. Series
2. DataFrame


### 1. Series
* A Pandas Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, Python objects, etc.).
* The axis labels are collectively called indexes.


### 2. DataFrame
* Pandas DataFrame is a two-dimensional data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

# 1. Series()

```python
## Seriees()

data = [1,2,3,4,5]

series = pd.Series(data)
print("series:\n",series)
```

```python
# creating a sample dataframe 
data = pd.DataFrame({
                    'Brand': ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Maruti', 'Hyundai', 'Renault', 'Tata', 'Maruti'], 
                    'Year': [2012, 2014, 2011, 2015, 2012, 2016, 2014, 2018, 2019], 
                    'Kms Driven': [50000, 30000, 60000, 25000, 10000, 46000, 31000, 15000, 12000], 
                    'City': ['Gurgaon', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Chennai',  'Ghaziabad'], 
                    'Mileage':  [28, 27, 25, 26, 28, 29, 24, 21, 24]
                    }) 
  
# displaying the DataFrame 
display(data)
```

```python
print(data['Brand'])

print()

print(type(data['Brand']))   ## type = series
```

## loc[ ]: *pass the index of the row or column which we want to select.* 

* **Note :** Last index includes the method

## 1. Selecting a Range of Rows From the DataFrame

```python
data.loc[0]  ## selecting single (0th) row 
```

```python
data.loc[3, ['Brand', 'Year']]  ## selecting single (3rd) row's -> ['Brand' and Year]column value
```

```python
## selecting a single value

data.loc[0, 'Brand']   ### 0th index ka "brand" name
```

```python
### selecting multiple rows (special rows)

data.loc[[0, 3]]
```

```python
## selecting multiple rows in sequence [start to end] :  end rows included

data.loc[0 : 5]
```

```python

display(data.loc[3:])  ## start from 3rd row to all row print

print(type(data.loc[3:]))  ## type = dataframe
```

```python
### selecting multiple rows in sequence by skipping rows

print(data.loc[1:6:2])  ## from 1st to 6th by skipping each 2nd rows of previous rows
```

## 2. Selecting Data According to Some Conditions

```python
# selecting cars with brand 'Maruti' and Mileage > 25 

data.loc[(data['Brand'] == 'Maruti') & (data.Mileage > 25)]
```

## 3. Updating the Value of Any Column

```python
# updating values of 'Mileage' if 'Year < 2015' 

data.loc[(data.Year < 2015), ['Mileage']] = 22 
display(data)
```

# iloc[ ] function
* position-based selecting method which means that we have to pass an integer index in the method to select a specific row/column.
* **Note :** This method does not include the last element of the range passed in it unlike loc()

```python
display(data.iloc[3:5])  ## from 3 to 4th row data print- Last index is not included unlike loc()
```

```python
## iloc : row index

# selecting 0th, 2nd, 4th, and 7th index rows 
data.iloc[[0, 2, 4, 7]]
```

```python
# selecting 'rows from 1 to 4' and c'olumns from 2 to 4 '

display(data.iloc[1: 5, 2: 5])
```

```python
data
```

```python
### select 1 to 3rd column value of 3rd rows

data.iloc[3, 1:4]
```

```python
## seelect all rows and all columns except last column

data.iloc[:, :-1]
```

## Accessing a specific element

```python
# creating position and label variables

### using at[]

position = 2  ## row
label = 'Brand'  ## column

display(data.at[position, label])
```

```python
### using iat[]  : indexing start fom 0

## access specific element using row and column index

display(data.iat[2,3])   ## 2nd row and 3rd column
```

```python
## Adding a column

data['Color'] = ['white','gray','red','black','white','blue','white','black','white']

data
```

```python
## Remove a column->  drop(col_name, axis=1, inplace=true)

## by default: inplace=Flase
## so the col will not remove permanetly, we have to make it True for Permanent delete
## by default: axis=0 (row delete): ---- axis=1 (col delete)

data.drop('Color', axis=1)  ## not permanently delete because 'inplace' attribute is by default Fallse
```

```python
display(data)
```

```python
## permanetly remove column

data.drop('Color', axis=1, inplace=True)
display(data)
```

```python
## simillarly permanetly remove a row

## axis=0 (by default) : for row deletion

data.drop(8, axis=0, inplace=True)  ## axis=0 is optional because is by default take it.
display(data)
```

```python
## add Mileage to 1

data["Mileage"] = data["Mileage"]+1

display(data)
```
